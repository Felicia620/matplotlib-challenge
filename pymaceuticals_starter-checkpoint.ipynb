## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata, study_results, on = "Mouse ID")

# Display the data table for preview
combined_data_df
```

```python
# Checking the number of mice

mouse_count = pd.DataFrame(
    
    {"Number of Mice": [combined_data_df["Mouse ID"].nunique()]})

mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint

duplicate_mice = combined_data_df[combined_data_df.duplicated(subset = ["Mouse ID", "Timepoint"])]

duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID

all_data_duplicate_mouse_id = combined_data_df[combined_data_df["Mouse ID"] == "g989"]

all_data_duplicate_mouse_id
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID

clean_combined_data = combined_data_df[combined_data_df["Mouse ID"] != "g989"]

clean_combined_data
```

```python
# Checking the number of mice in the clean DataFrame

mouse_count = pd.DataFrame(
    
    {"Number of Mice": [clean_combined_data["Mouse ID"].nunique()]})

mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM 
# of the tumor volume for each regimen

regimen_mean = clean_combined_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]

regimen_median = clean_combined_data.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]

regimen_variance = clean_combined_data.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]

regimen_standard_deviation = clean_combined_data.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]

regimen_standard_error_mean = clean_combined_data.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume 
# Assemble the resulting series into a single summary dataframe

summary_table_df = pd.DataFrame(
    
    {"Mean": regimen_mean,
     
     "Median": regimen_median,
     
     "Variance": regimen_variance,
     
     "Standard Deviation": regimen_standard_deviation,
     
     "Standard Error of the Mean": regimen_standard_error_mean})

summary_table_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

aggregate_summary_table = clean_combined_data.groupby("Drug Regimen").agg(["mean",
                                                                           "median",
                                                                           "var",
                                                                           "std",
                                                                           "sem"])["Tumor Volume (mm3)"]

aggregate_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

last_timepoint = clean_combined_data.groupby(["Mouse ID"])["Timepoint"].max()

last_timepoint_df = last_timepoint.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

tumor_volume_merge = pd.merge(last_timepoint_df, clean_combined_data, on = ("Mouse ID", "Timepoint"))

tumor_volume_merge
```

```python
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 

capomulin_df = clean_combined_data.loc[clean_combined_data["Drug Regimen"] == "Capomulin", :]

capomulin_last_timepoint = capomulin_df.groupby("Mouse ID").max()["Timepoint"]

capomulin_tumor_volume = pd.DataFrame(capomulin_last_timepoint)

capomulin_merge = pd.merge(capomulin_tumor_volume, clean_combined_data, on = ("Mouse ID", "Timepoint"))

capomulin_merge
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Determine outliers using upper and lower bounds

capomulin_tumor_volume = capomulin_merge["Tumor Volume (mm3)"]

capomulin_quartiles = capomulin_tumor_volume.quantile([0.25, 0.5, 0.75])

capomulin_lower_quartile = capomulin_quartiles[0.25]

capomulin_upper_quartile = capomulin_quartiles[0.75]

capomulin_interquartile_range = capomulin_upper_quartile - capomulin_lower_quartile

capomulin_lower_bound = capomulin_lower_quartile - (1.5 * capomulin_interquartile_range)

capomulin_upper_bound = capomulin_upper_quartile + (1.5 * capomulin_interquartile_range)

print(f"The interquartile range of Capomulin tumors is {capomulin_interquartile_range}.")

print(f"Values below {capomulin_lower_bound} are outliers.")

print(f"Values above {capomulin_upper_bound} are outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

capomulin_tumor_volume = capomulin_merge["Tumor Volume (mm3)"]

fig1, ax1 = plt.subplots()

ax1.set_title("Capomulin Regimen")

ax1.set_ylabel("Final Tumor Volume")

ax1.boxplot(capomulin_tumor_volume)

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation_coefficient = round(st.pearsonr(capomulin_average["Weight (g)"], capomulin_average["Tumor Volume (mm3)"])[0], 2)

print(f"The correlation between mouse weight and average tumor volume for the Capomulin regimen is {correlation_coefficient}.")
```

```python
# Printed equation manually in cell below, plt.annotate did not work

x_values = capomulin_average["Weight (g)"]

y_values = capomulin_average["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))

plt.scatter(x_values,y_values)

plt.plot(x_values, regress_values,"r")

plt.figtext(.55, .35, "y = 0.95x + 21.55", fontsize = 15, color = "red")

plt.xlabel("Weight (g)")

plt.ylabel("Tumor Volume (mm3)")

plt.show()
```
